# Regressão Polinomial

Um exemplo de split teste e predict na Regressao Polinomial

```python
#Importe as bibliotecas que irá usar

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
# TODO: Crie a vaíavel do preditor e de saída. Carregue os dados
train_data = pd.read_csv('https://raw.githubusercontent.com/WoMakersCode/data-science-bootcamp/master/3.%20Modelos%20regressivos/Exercicio_3/position_salaries.csv')
X = train_data.iloc[:, 1:2].values
y = train_data.iloc[:, 2].values
```

```python
#Split test
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.2)
```

```python
#Criando Regressao Linear

lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Visualizando resultados
def viz_linear():
    plt.scatter(X, y, color='red')
    plt.plot(X, lin_reg.predict(X), color='blue')
    plt.title('Truth or Bluff (Linear Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
    return
viz_linear()
```

```python
# Criando a regressao Polinomial

poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)

# Visualizando resultados
def viz_polymonial():
    plt.scatter(X, y, color='red')
    plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color='blue')
    plt.title('Truth or Bluff (Linear Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
    return
viz_polymonial()
```

```python
#Fazendo previsao

# Fazendo previsão
lin_reg.predict([[5.5]])


# Fazendo previsao com Polymonial Regression
pol_reg.predict(poly_reg.fit_transform([[5.5]]))
```
